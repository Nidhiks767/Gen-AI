```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mdickyriandi/crime-incident")

print("✅ Path to dataset files:", path)
```

```python
import os

files = os.listdir(path)
print("✅ Files in dataset folder:")
print(files)
```

```python
import os
import pandas as pd

# List all files downloaded
all_files = os.listdir(path)
print("✅ Files found:", all_files)

# Find Excel file
xlsx_files = [f for f in all_files if f.endswith(".xlsx")]
print("✅ Excel files found:", xlsx_files)

file_path = os.path.join(path, xlsx_files[0])
print("✅ Using file:", file_path)
```

```python
xls = pd.ExcelFile(file_path)
print("✅ Sheets available in Excel:", xls.sheet_names)
```

```python
df_narkotika = pd.read_excel(file_path, sheet_name="Narkotika")
print("✅ Narkotika Loaded:", df_narkotika.shape)
df_narkotika.head()
```

```python
print("✅ Columns in dataset:\n")
print(df_narkotika.columns)

print("\n✅ Dataset Info:\n")
df_narkotika.info()

print("\n✅ Missing Values (Top):\n")
print(df_narkotika.isnull().sum().sort_values(ascending=False))
```

```python
print("✅ Summary Statistics:\n")
df_narkotika.describe()
```

```python
print("✅ Duplicate Rows:", df_narkotika.duplicated().sum())
```
