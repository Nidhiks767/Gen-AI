```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# House size (input feature)
X = np.array([600, 800, 1000, 1200], dtype=float)

# House price (target)
y = np.array([3, 4, 6, 7], dtype=float)
```

```python
X = X / 1000   # scale values between 0 and 1
```

```python
[0.6, 0.8, 1.0, 1.2]
```

```python
w = 0.0   # weight
b = 0.0   # bias
```

```python
learning_rate = 0.01
epochs = 1000
```

```python
def predict(X, w, b):
    return w * X + b
```

```python
def compute_loss(y_true, y_pred):
    return np.mean((y_pred - y_true) ** 2)
```

```python
n = len(X)

for epoch in range(epochs):

    # Step 1: Predict
    y_pred = predict(X, w, b)

    # Step 2: Calculate gradients
    dw = (2/n) * np.sum((y_pred - y) * X)
    db = (2/n) * np.sum(y_pred - y)

    # Step 3: Update parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db

    # Step 4: Print loss
    if epoch % 100 == 0:
        loss = compute_loss(y, y_pred)
        print(f"Epoch {epoch}, Loss = {loss}")
```

```python
print("Final weight:", w)
print("Final bias:", b)
```

```python
new_size = 900 / 1000
predicted_price = predict(new_size, w, b)

print("Predicted price:", predicted_price)
```

```python
plt.scatter(X, y, label="Actual Data")
plt.plot(X, predict(X, w, b), label="Model Prediction")

plt.xlabel("House Size (normalized)")
plt.ylabel("Price")
plt.legend()
plt.show()
```
